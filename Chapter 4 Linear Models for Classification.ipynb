1.hyperplane, 超平面


超平面：在一个D维Euclidean  space中的（超）平面是一它的一个D-1维流形，而且该空间是一个线性空间。    Linearly separable：分布于D维空间中的全部数据点可以用超平面无错地分隔成类Coding scheme：1-of-K binary coding scheme，即如果有K个类，某数据点属于第i个类，则表示为一个K维向量，该向量除了第i个分量是1，其余都是0。Feature vector：对于一个D维input x，使用一个fixed nonlinear transformation将其映射成同样是D维的向量，即feature vector。
